```python
import torch
import os

from huggingface_hub import HfApi, login
```

```python
checkpoint_path = "/content/drive/Shareddrives/_PercobaanKaenova/Tugas Akhir/checkpoints"
```

```python
# Resave pytorch non cpu to cpu
indobertweet_gpu_checkpoint = f"{checkpoint_path}/indobertweet_gpu"
indobertweet_checkpoint = f"{checkpoint_path}/indobertweet"

checkpoints = os.listdir(indobertweet_gpu_checkpoint)
for checkpoint in checkpoints:
    data = torch.load(f"{indobertweet_gpu_checkpoint}/{checkpoint}", map_location="cpu")
    torch.save(data, f"{indobertweet_checkpoint}/{checkpoint}")
```
